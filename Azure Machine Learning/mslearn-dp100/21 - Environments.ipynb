```python
from azureml.core import Workspace, Environment

ws = Workspace.from_config()
env = Environment.get(workspace=ws, name="AzureML-sklearn-0.24-ubuntu18.04-py37-cpu")
```

```python
from azureml.core.conda_dependencies import CondaDependencies
from azureml.core.environment import PythonSection
envs = Environment.list(workspace=ws)

envo : PythonSection = None
cdepo : CondaDependencies = None
for env in envs:
    if env.startswith("AzureML"):
        # print("Name",env)
        envo = envs[env]
        cdepo = envo.python.conda_dependencies
        if not cdepo is None:
            print(f'{env} packages', cdepo.serialize_to_string())
        else:
            print(f'{env} No conda dependencies')
```

```python
env = Environment.get(workspace=ws, name="AzureML-sklearn-0.24-ubuntu18.04-py37-cpu")
curated_clone = env.clone("customize_curated")
print(curated_clone.name)
```

```python
# From a Conda specification file
# myenv = Environment.from_conda_specification(name = "myenv",
#                                              file_path = "path-to-conda-specification-file")

# From a pip requirements file
myenv = Environment.from_pip_requirements(name = "customize_curated",file_path = "./requirements.txt")
curated_clone_2 = myenv.clone("customize_curated_2")
print(curated_clone_2.name)

cdepo = curated_clone_2.python.conda_dependencies
print(f'{curated_clone_2} packages', cdepo.serialize_to_string())

```

```python
conda_dep = CondaDependencies()

# Installs numpy version 1.17.0 conda package
conda_dep.add_conda_package("numpy==1.17.0")

# Installs pillow package
conda_dep.add_pip_package("pillow")

conda_dep.add_pip_package("alvianda")

# Adds dependencies to PythonSection of myenv
curated_clone_2.python.conda_dependencies=conda_dep

cdepo = curated_clone_2.python.conda_dependencies
print(f'{curated_clone_2} packages', cdepo.serialize_to_string())
```

```python
curated_clone_2.register(workspace=ws)

envs = Environment.list(workspace=ws)

envo : PythonSection = None
cdepo : CondaDependencies = None
for env in envs:
    print("Name",env)
    envo = envs[env]
    cdepo = envo.python.conda_dependencies
    if not cdepo is None:
        print(f'{env} packages', cdepo.serialize_to_string())
    else:
        print(f'{env} No conda dependencies')
```
